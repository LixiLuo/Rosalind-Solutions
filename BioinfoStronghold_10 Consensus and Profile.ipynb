```python
path='C:\\Users\Luo_L\Desktop\Python Examples/rosalind_cons.txt'
f=open(path).read()
f1=f.replace('\n','')
f2=f1.split('>')
f2
```

```python
fx=[]
for xx in f2[1:]:
    ff=xx[13:]
    fx.append(ff)
fx
```

```python
fy=[]
for y in fx:
    fy.append(list(y))
fy
#np.array
```

```python
lenth=len(fy[0])
```

```python
cA=[]
cC=[]
cG=[]
cT=[]
for n in range(0,lenth):
    columnA=0
    columnC=0
    columnG=0
    columnT=0
    for x in fy:
        if x[n]=='A':
            columnA+=1
        elif x[n]=='C':
            columnC+=1        
        elif x[n]=='G':
            columnG+=1
        elif x[n]=='T':
            columnT+=1
    cA.append(columnA)
    cC.append(columnC)
    cG.append(columnG)
    cT.append(columnT)
```

```python
Dict=dict(zip(['A','C','G','T'],[cA,cC,cG,cT]))
Dict
```

```python
List=[]
for N in range(0,len(Dict['A'])):
    if Dict['A'][N]==max(Dict['A'][N],Dict['C'][N],Dict['G'][N],Dict['T'][N]):
        List.append('A')
    elif Dict['C'][N]==max(Dict['A'][N],Dict['C'][N],Dict['G'][N],Dict['T'][N]):
        List.append('C')
    elif Dict['G'][N]==max(Dict['A'][N],Dict['C'][N],Dict['G'][N],Dict['T'][N]):
        List.append('G')
    elif Dict['T'][N]==max(Dict['A'][N],Dict['C'][N],Dict['G'][N],Dict['T'][N]):
        List.append('T')
List
```

```python
for z in List:
    print(z,end='')
for nn in 'ACGT':
    print('\n'+nn+': ',end='')
    for zz in Dict[nn]:
        print(zz,end=' ')
#listmap(str,list)
```
